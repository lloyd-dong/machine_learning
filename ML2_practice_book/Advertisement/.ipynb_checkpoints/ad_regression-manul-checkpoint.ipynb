回归问题的评价测度
对于分类问题，评价测度(evaluation metrics)是准确率，但这种方法不适用于回归问 题。我们使用针对连续数值的评价测度。
这里介绍 3 种常用的针对线性回归的测度。 
* 平均绝对误差(Mean Absolute Error, MAE) 
* 均方误差(Mean Squared Error, MSE) 
* 均方根误差(Root Mean Squared Error, RMSE)

```python
def cal_RMSE(predict, test):
    #type(test) is pandas.core.series.Series, tyep(predict) is np.array
    sum_mean=0
    for i in range(len(predict)):
        sum_mean += (predict[i] - test.values[i])**2  
    print 'RMSE by hand is ', np.sqrt(sum_mean/len((predict)))
```

```python
def plot_compare(predict,test):
    plt.figure(figsize=(6,4))
    plt.plot(range(len(predict)), predict,'b', label='predict')
    plt.plot(range(len(predict)),test,'r',label='test')
    plt.legend(loc='upper right')
    plt.xlabel('the number of sales')
    plt.ylabel('value of sales')
    plt.show()
```

```python
plot_compare(y_pred,y_test)
```

```python
cal_RMSE(y_pred,y_test)
```

Dataframe['abc'] 的类型是 <class 'pandas.core.series.Series'>, shape 是 (200L,) 一维数组
Dataframe[['abc']] 的类型是 <class 'pandas.core.frame.DataFrame'>,shape (200, 1), 只有一列的data frame

如果y 目标用这两种不同的类型表示，线性模型参数(model.intercept_, coef_的值是一样的，但是类型不同，用起来会有些麻烦

```python
def train_predict(features,target_label,data):
    assert(isinstance(target_label, str)) #please refer the document above
    X=data[features]
    Y=data[target_label]
    X_train,X_test,Y_train,Y_test= train_test_split(X,Y,random_state=1)
    linreg=LinearRegression()
    model=linreg.fit(X_train, Y_train)
    print model
    print model.intercept_, type(linreg.intercept_)
    print model.coef_, type(linreg.coef_),linreg.coef_.shape    
    print(zip(features,linreg.coef_))
    Y_pred=model.predict(X_test)
    cal_RMSE(Y_pred, Y_test)
    return Y_pred, Y_test
```

```python
features=['TV','Radio','Newspaper']
target='Sales'
y_pred, y_test=train_predict(features,target,data)
plot_compare(y_pred,y_test)

features=['TV','Radio']
y2_pred, y_test=train_predict(features,target,data)
plot_compare(y_pred,y2_pred)
```

$\Theta^T \cdot X - Y $

展开 $\Theta : \theta_1, \theta_2, ... \theta_j,..., \theta_p $

$$  \left( \begin{array}{ccc}
X_1: & (X_11, X_12, ... X_1p) & Y_1 \\
X_2: & (X_21, X_22, ... X_2p) & Y_2 \\
...  \\
X_n: & (X_n1, X_n2, ... X_np) & Y_n \end{array} \right) $$
  
$$J(\Theta) = \frac{1}{2n}\sum_i^n(\Theta \cdot X_i -Y_i)^2 $$

$$ \frac{\partial J(\Theta)}{\partial \Theta} =   \left\{ \frac{\partial J(\Theta)}{\theta_1}, \frac{\partial J(\Theta)}{\theta_2}, ...,\frac{\partial J(\Theta)}{\theta_j}..., \frac{\partial J(\Theta)}{\theta_p} \right\} $$
