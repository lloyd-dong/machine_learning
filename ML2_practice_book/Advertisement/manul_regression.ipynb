```python
import pandas as pd
import numpy as np
from sklearn.cross_validation import train_test_split

#x, y 是 np.ndarray
def regression(x,y, alpha):
    theta = np.zeros(x.shape[1])
    for times in range(100):
        m1=map(lambda x_i, y_i: np.dot(theta,x_i)- y_i, x,y ) #得到数组 [theta*x[1]-y[1], theta*x[2]- y[2], ..., theta * x[n]-yn], 1维 长度为n
        m2=[np.dot(m1, x_t) for x_t in x.T]  # x_t 是 x.T 的每一行，也就是原来x 的每一列，np.dot(m1, x_t) 就是 [attach]4776[/attach]
        theta = theta-[alpha *m for m in m2]
        print times, theta
    return theta

#用广告的数据做回归
data=pd.read_csv("Advertising.csv")
X=data[['TV','Radio']]
Y=data['Sales']
x_train, x_test, y_train, y_test = train_test_split(X,Y,random_state=1)

n=x_train.shape[0]
x_train.loc[:,'theta0'] =pd.Series(np.ones(n),index=x_train.index) # 给x_train 增加一列 1
regression(x_train.values, y_train.values, 0.01) 
```
