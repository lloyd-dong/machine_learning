```python
import pandas as pd
import numpy as np
from sklearn.cross_validation import train_test_split
```

```python
#用广告的数据做回归
data=pd.read_csv("Advertising.csv")
X=data[['TV','Radio']]
Y=data['Sales']
x_train, x_test, y_train, y_test = train_test_split(X,Y,random_state=1)
```

```python
#x, y 是 np.ndarray
def regression1(x,y, alpha):
    theta = np.zeros(x.shape[1])
    for times in range(100):
        m1=map(lambda x_i, y_i: np.dot(theta,x_i)- y_i, x,y ) #得到数组 [theta*x[1]-y[1], theta*x[2]- y[2], ..., theta * x[n]-yn], 1维 长度为n
        m2=[np.dot(m1, x_t) for x_t in x.T]  # x_t 是 x.T 的每一行，也就是原来x 的每一列，np.dot(m1, x_t) 就是 [attach]4776[/attach]
        theta = theta-[alpha *m for m in m2]
        print times, theta
    return theta
```

```python
def regression2(x,y,alpha):
    n,p =x.shape   
    theta = np.zeros(p)
    for times in range(100):
        t_bar=[]
        m1=map(lambda x_i, y_i: np.dot(theta,x_i)- y_i, x,y )        
        for j in range(p): #求每一个thta
            g=0.0
            for i in range(n):
                g +=m1[i]*x[i][j]
            g = g/n
            #print "g", g
            t_bar.append(theta[j] - alpha * g)
        theta=t_bar
        print theta
    return theta
```

```python
def regression3(x,y,alpha):
    n,p =x.shape   
    theta = np.zeros(p)
    for times in range(100):
        t_bar=[]
        for j in range(p): #求每一个thta
            g=0.0
            for i in range(n):
                g += (np.dot(theta,x[i])-y[i])*x[i][j]
            g = g/n
            #print "g", g
            t_bar.append(theta[j] - alpha * g)
        theta=t_bar
        print theta
    return theta
```

```python
def regression(data, alpha,lamda):
    n=len(data[0])-1 # 去掉y, target列
    theta= np.zeros(n)
    for times in range(100):
        for d in data:
            x = d[:-1]
            y = d[-1]
            g = np.dot(theta, x) -y
            theta = theta -alpha * g * x + lamda * theta
        print times, theta
    return theta
```
